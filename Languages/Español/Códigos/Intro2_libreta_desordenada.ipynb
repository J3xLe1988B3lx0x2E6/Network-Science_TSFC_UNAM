```python
N = 1000
k = 10
ws0 = nx.watts_strogatz_graph(N,k,0)
#Caminos más cortos
l0 = nx.average_shortest_path_length(ws0)
#ClusteringCoef
c0 = nx.average_clustering(ws0)
```

```python
ws = nx.watts_strogatz_graph(N,k,0.05)
```

```python
nx.draw_circular(ws)
```

```python
C = nx.average_clustering(ws)
print(C/c0)
l = nx.average_shortest_path_length(ws)
print(l/l0)
```

```python
import matplotlib.pyplot as plt
import numpy as np

n = 14
p = np.logspace(0,1,n)
s = np.linspace(0,1)
print(p)
C = []
L = []

for i in range(n):
    ws = nx.watts_strogatz_graph(N,k,p[i])
    c_n = nx.average_clustering(ws)
    l_n = nx.average_shortest_path_length(ws)
    C.append(c_n/c0)
    L.append(l_n/l0)
    
fig, ax = plt.subplots()
ax.plot(s.reshape(0,50),C.reshape(0,1),"ro")
ax.plot(s.reshape(0,50),L.reshape(0,1),"g--")
plt.xscale("log")
plt.show()
```
