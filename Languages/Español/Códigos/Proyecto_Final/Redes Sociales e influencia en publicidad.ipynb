```python
import networkx as nx
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import os
import graphsnx as gf
```

```python
path = os.path.abspath(os.getcwd())
ruta_red_1 = r'{}/Data_reference/rt-retweet/rt-retweet.mtx'.format(path)
ruta_red_2 = r'{}/Data_reference/socfb-nips-ego/socfb-nips-ego.edges'.format(path)
ruta_red_3 = r'{}/Data_reference/web-spam/web-spam.mtx'.format(path)
Tt = nx.read_edgelist(ruta_red_1,nodetype=int,create_using=nx.Graph())
Fb = nx.read_edgelist(ruta_red_2,nodetype=int,create_using=nx.Graph())
Wb = nx.read_edgelist(ruta_red_3,nodetype=int,create_using=nx.Graph())
```

```python
#Caracteristicas de la red.
def Max_minDegree(graph):
    deg = dict(nx.degree(graph))
    lista = []
    for i in deg:
        lista.append(deg[i])
    prom = sum(lista)/nx.number_of_nodes(graph)
    return min(lista), max(lista), prom #Maximo Degree y Min degree,promedio
min_degree_t,max_degree_t, prom_t = Max_minDegree(Tt)
min_degree_f,max_degree_f, prom_f = Max_minDegree(Fb)
min_degree_w,max_degree_w, prom_w = Max_minDegree(Wb)
    
Infor_red = {
    1:{
        "Nombre":"Twitter_Retweet",
        "Nodos":nx.number_of_nodes(Tt),
        "Aristas":nx.number_of_edges(Tt),
        "Pesada":nx.is_weighted(Tt),
        "Dirigida":nx.is_directed(Tt),
        "MaxDegree":max_degree_t,
        "MinDegree":min_degree_t,
        "PromDegree":prom_t,
    },
    2:{
        "Nombre":"Social Friendship Network (Facebook)",
        "Nodos":nx.number_of_nodes(Fb),
        "Aristas":nx.number_of_edges(Fb),
        "Pesada":nx.is_weighted(Fb),
        "Dirigida":nx.is_directed(Fb),
        "MaxDegree":max_degree_f,
        "MinDegree":min_degree_f,
        "PromDegree":prom_f,
    },
    3:{
        "Nombre":"Web Spam",
        "Nodos":nx.number_of_nodes(Wb),
        "Aristas":nx.number_of_edges(Wb),
        "Pesada":nx.is_weighted(Wb),
        "Dirigida":nx.is_directed(Wb),
        "MaxDegree":max_degree_w,
        "MinDegree":min_degree_w,
        "PromDegree":prom_w,
    }
}
print("\t Caracteristicas de las redes: \n")
for x in Infor_red:
    print("Nombre: "+Infor_red[x]['Nombre']+"\nNodos: "+str(Infor_red[x]['Nodos'])+"\nAristas: "+str(Infor_red[x]['Aristas'])+"\nPesada? "+str(Infor_red[x]['Pesada'])+"\nDirigida? "+str(Infor_red[x]['Dirigida'])+"\nMax. Degree: "+str(Infor_red[x]["MaxDegree"])+"\nMin. Degree: "+str(Infor_red[x]["MinDegree"])+"\nProm. Degree: "+str(Infor_red[x]["PromDegree"])+"\n")
```

## Clustering Coefficient. üßÆüòÄ‚û°Ô∏èüôÖüèªüôçüèª‚Äç‚ôÇÔ∏è

Como recordaremos el *Clustering Coefficient* es cuantifica qu√© tanto est√° de agrupado (o interconectado) con sus vecinos (o nodos juntos).

\begin{equation}
    C_{i} = \frac{~numero~ ~de ~triangulos~ conectados~}{~numero~ de~triangulos~posibles~}
\end{equation}
Entonces matem√°ticamente se representa como:

\begin{equation}
    \displaystyle C_{i} = \frac{A^{3}}{\binom{k_{i}}{2}}
\end{equation}
Donde $A$ es la matriz de adjyacencia.

En este caso vamos a comparar el *Clustering Coefficient* tanto para la red de *Twitter*, *Web Spamming*  y de *Facebook*. Ya que nos dar√° informaci√≥n de tan conectado estar√° cada red y as√≠ tener un pronostico si alguien comparte un v√≠deo o tweet, ser√≠a el medio el cual tendr√≠a m√°s alcance.


```python
#Matriz de adjyacencia 
m_tt = nx.to_numpy_array(Tt)
m_wp = nx.to_numpy_array(Wb)
m_fb = nx.to_numpy_array(Fb)
print("\tTwitter MA:\n {}".format(m_tt))
print("\tWeb Spamming MA:\n{}".format(m_wp))
print("\tFacebook MA:\n{}".format(m_fb))
```

```python
#Para Twitter el promedio de Clustering Coeffient.
clc_tt = nx.average_clustering(Tt)
#Para YouTube el promedio de Clustering Coefficient.
clc_fb = nx.average_clustering(Fb)
#Para SpamWeb el promedio de Clustering Coefficient.
clc_wb = nx.average_clustering(Wb)
print("\t                         Valores de Clustering Coefficient \t")
print("\t|------------------------------------------------------------------------|")
print("\t| Network  | Average Clustering Coefficient | Comparacion de redes       |")
print("\t| -------- | ------------------------------ | ---------------------------|")
print("\t| Facebook | {}           | Facebook < WS: {}        |".format(clc_fb,clc_fb<clc_wb))
print("\t| Twitter  | {}           | Twitter < Facebook: {}  |".format(clc_tt,clc_tt<clc_fb))
print("\t| WS       | {}             | WS < Twitter: {}        |".format(clc_wb,clc_wb<clc_tt))
print("\t| -------- | ------------------------------ | Fb < Tt < WS: {}         |".format(clc_fb<clc_tt<clc_wb))
```

## Betweenness centrality.üèπüéØ‚≠ï‚¨ÖÔ∏è

Se conoce como *centrailidad de intermediaci√≥n* en el cual un nodo $i$ es el n√∫mero de la distancia entre dos v√©rtices o nodos de un grafo a la longitud o n√∫mero de entre dos v√©rtices o nodos de un grafo a la longitud o n√∫mero de aristas del camino m√°s corto entre ellos, i.e. $i$ representa el n√∫mero de *caminos geod√©sicos*, es decir que el *betweenness centrality* $\beta_{i}$ de un nodo $i$ es:

\begin{equation}
    \beta_{i} = \sum_{j,k}\frac{\Lambda_{jk}(i)}{\Lambda_{jk}}
\end{equation}

Donde $\displaystyle\Lambda_{jk}$ es el n√∫mero total de caminos geod√©sicos entre cualesquiera dos nodos $j$ y $k$, mientras que $\Lambda_{j,k}(i)$ es el n√∫mero caminos geod√©sicos entre los nodos $j$ y $k$ que adem√°s pasan a trav√©z del nodo $i$.


En este caso se quiere recuperar los nodos con m√°s conexiones que pasa sobre ellos de cada una de las redes y se graficar√° los nodos m√°s importantes en la red.

## Network Efficiency. ‚§¥Ô∏èüìâüåê‚ùå


La eficiencia de una red es la medida entre nodo y los dem√°s $\epsilon_{i}$ y mide que tan eficiente puede fluir informaci√≥n en este caso ser√° sobre las redes sociales. En la cual se define como: $\displaystyle\epsilon_{ij} = \frac{1}{d_{ij}}$ donde $d_{ij}$ es la longitud del camino m√°s corto entre $i$ y $j$. Si la distancia entre ambos nodos es grande la *eficiencia* es poca y viceversa. De √©sta manera, la *eficiencia* promedio del nodo $i$ viene dada por:

\begin{equation}
    \langle\epsilon_{i}\rangle = \frac{1}{n-1}\sum_{j\not=i}\frac{1}{d_{ij}}
\end{equation}

En este caso se va analizar la eficiencia de la red pero especialmente entre los nodos que ten√≠an mayor Betweennes en la red para ver que tan eficiente son en esos nodos, tambi√©n para hacer un diagrama entre la eficiencia de aquellos nodos y entre cuento estar√° distribuida.

<h2> Algoritmo de <b>Girvan - Newman.</b></h2>

Este fue uno de los primeros algoritmos propuestos por Girvan y Newman para la busqueda de comunidades o la detecci√≥n de las mismas, que en donde las aritas se seleccionan de acuerdo a los volares de las medidas de *intermedic√≥n de aristas* (***edge betweenness***). estimando la importancia de las aristas. La forma de explicar este algorimo de manera como pseudoc√≥digo es:

1. C√°lculo de la intermediaci√≥n para todas las aristas.
2. Eliminaci√≥n de la arita con mayor centralidad.
3. Nuevo calculo de la itermediaci√≥n.
4. Se repite el paso 2.

Como se puede pareciar ser un proceso recursivo por lo cual se puede hacer tardado en algunas ocaciones, para ello en la red de *karate* se puede apreciar las comunidades y las particiones mediante de un *dendograma*. 

En este caso nada m√°s vamos a probar para la red de *Twitter* hacer su dendograma y apreciar sus particiones, y se dar√° la explicaci√≥n del por qu√© no se hizo lo mismo en la red de *Facebook* y *Web spam*.

<h2> Eficiencia vs el algoritmo de Girvan - Newman</h2>

Cierto que el algorimo de ***Girvan - Newman*** es uno de los algorimos de busqueda de comunidades pero tenemos un peque√±o problema. 

El problema es tiempo de ejecuci√≥n. El an√°lisis que se le hizo al algorimo su complejidad fue  $O(nm^{2})$. Es una big-$O$ no lineal sino cuadratico. Por lo tanto, para las redes de *Facebook* y *Twitter* el tiempo de ejecuci√≥n para hacer el dendograma era demasiado.

Sin embargo, existe otro algorimo que nos permite encontrar las particines de nuestra red con mayor presici√≥n y vamos a comparar el algorimo de ***Girvan - Newman*** contra el algoritmo de ***Louvain***. Adem√°s de obtener sus particines veremos cual es la mejor.

<h2> Algoritmo de Louvain</h2>

En primera estancia consiste en un *barrido* secuancial sobre todos los nodos, dado un nodo $i$, se calcula la ganancia de la modularidad, que proviene de poner $i$ en la comunidad de su vecino y seleccionar la comunidad del vecino que produce el mayor incremento de $Q_{w}$ siempre que sea positivo. Esto es importante ya que este algoritmo su complejidad es de $O(m)$ a $O(n\log{\langle k\rangle})$ cuando se tiene redes con un estructura limpia en comunidades. 

```python
# Tiempo de ejecuci√≥n y mejor particion.
#Twitter.Girvan -Newman.
import time
inicio = time.time()
communities_generator = nx.community.girvan_newman(Tt)
top_level_communities = next(communities_generator)
next_level_communities = next(communities_generator)
part_tt = sorted(map(sorted, next_level_communities))
print(nx.community.is_partition(Tt,part_tt))
print("Aproximaci√≥n de mejor particion: ",nx.community.modularity(Tt,part_tt))
fin = time.time()
tem_tt = fin-inicio
print("Tiempo total (Red de Twitter): {}".format(tem_tt))
```

```python
# Tiempo de ejecuci√≥n y mejor particion.
#Facebook. Girvan -Newman.

inicio = time.time()
communities_generator = nx.community.girvan_newman(Fb)
top_level_communities = next(communities_generator)
next_level_communities = next(communities_generator)
part_fb = sorted(map(sorted, next_level_communities))
print(nx.community.is_partition(Fb,part_fb))
print(nx.community.modularity(Fb,part_fb))
fin = time.time()
tem_fb = fin-inicio
print("Tiempo total (Red de Facebook): {}".format(tem_fb))
```

<h3> Resultados. </h3>
En este caso la comparaci√≥n entre la red de Twitter fue m√°s r√°pida su ejecuci√≥n ya que la red no est√° tan pesada. Sin embargo, como se puede apreciar en alre de Facebook, fue mucho m√°s el timpo. Por lo cual implic√≥ m√°s tiempo. *time.time()* toma el tiempo en segundos y microsegundos para medidas de procesamiento. En este caso se tard√≥ mucho m√°s. En el caso de *web spam* todav√≠a seguir√≠a ejecutandose ya que el algorimo de **Girvan - Newman** es recursivo. 

En el caso del algorimo de ***Louvain*** las particines son m√°s precisas a comparaci√≥n del anterior algoritmo. En este caso s√≠ se dar√° el an√°lisis para las 3 redes. 

```python
#Algorimo de Louvain en Twitter

inicio = time.time()
tt_l = nx.community.louvain_communities(Tt)
print(nx.community.is_partition(Tt,tt_l))
print("Aproximaci√≥n de mejor particion: ",nx.community.modularity(Tt,tt_l))
fin = time.time()
tem_tt = fin-inicio
print("Tiempo total (Red de Twitter): {}".format(tem_tt))
```

```python
#Algorimo de Louvain en Facebook

inicio = time.time()
fb_l = nx.community.louvain_communities(Fb)
print(nx.community.is_partition(Fb,fb_l))
print("Aproximaci√≥n de mejor particion: ",nx.community.modularity(Fb,fb_l))
fin = time.time()
tem_fb = fin-inicio
print("Tiempo total (Red de Twitter): {}".format(tem_fb))
```

```python
#Algorimo de Louvain en Web spam

inicio = time.time()
wb_l = nx.community.louvain_communities(Wb)
print(nx.community.is_partition(Wb,wb_l))
print("Aproximaci√≥n de mejor particion: ",nx.community.modularity(Wb,wb_l))
fin = time.time()
tem_wb = fin-inicio
print("Tiempo total (Red de Web Spam): {}".format(tem_wb))
```
